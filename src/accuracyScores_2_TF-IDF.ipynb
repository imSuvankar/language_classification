```python
import pandas
data = pandas.read_csv("/content/gdrive/MyDrive/Colab Notebooks/datasets/3_genetic_sequencing.csv")
x = data["seq"]
y = data["id"]

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
x = tfidf.fit_transform(x)

from sklearn.preprocessing import LabelEncoder
lblEnc = LabelEncoder()
y = lblEnc.fit_transform(y)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

#### 1. Using `Multinomial NB`

```python
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(x_train, y_train)

y_pred_MNB = MNB.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"MNB & TF-IDF: {accuracy_score(y_test, y_pred_MNB) * 100 :.2f}%")
```

#### 2. Using `Support Vector Machines (SVM)`

```python
from sklearn.svm import SVC
SVM = SVC()
SVM.fit(x_train, y_train)

y_pred_SVM = SVM.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"SVM & TF-IDF: {accuracy_score(y_test, y_pred_SVM) * 100 :.2f}%")
```

#### 3. Using `Random Forest`

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(x_train, y_train)

y_pred_RF = RF.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Random Forest & TF-IDF: {accuracy_score(y_test, y_pred_RF) * 100 :.2f}%")
```

#### 4. Using `Gradient Boosting Classifier`

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier()
GB.fit(x_train, y_train)

y_pred_GB = GB.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Gradient Boosting Classifier & TF-IDF: {accuracy_score(y_test, y_pred_GB) * 100 :.2f}%")
```

#### 5. Using `Decision Tree Classifier`

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)

y_pred_DT = DT.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Decision Tree Classifier & TF-IDF: {accuracy_score(y_test, y_pred_DT) * 100 :.2f}%")
```

#### 6. Using `K-Nearest Neighbors Classifier`

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
KNN.fit(x_train, y_train)

y_pred_KNN = KNN.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"K-Nearest Neighbors Classifier & TF-IDF: {accuracy_score(y_test, y_pred_KNN) * 100 :.2f}%")
```

#### 7. Using `AdaBoost Classifier`

```python
from sklearn.ensemble import AdaBoostClassifier
ADB = AdaBoostClassifier()
ADB.fit(x_train, y_train)

y_pred_ADB = ADB.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"AdaBoost Classifier & TF-IDF: {accuracy_score(y_test, y_pred_ADB) * 100 :.2f}%")
```

#### 8. Using `Logistic Regression`

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train, y_train)

y_pred_LR = LR.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Logistic Regression & TF-IDF: {accuracy_score(y_test, y_pred_LR) * 100 :.2f}%")
```

#### 9. Using `Extra Trees Classifier`

```python
from sklearn.ensemble import ExtraTreesClassifier
ET = ExtraTreesClassifier()
ET.fit(x_train, y_train)

y_pred_ET = ET.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Extra Trees Classifier & TF-IDF: {accuracy_score(y_test, y_pred_ET) * 100 :.2f}%")
```

#### 10. Using `Gaussian Process Regressor`

```python
x_train = x_train.toarray()
x_test = x_test.toarray()

from sklearn.gaussian_process import GaussianProcessRegressor
GPR = GaussianProcessRegressor()
GPR.fit(x_train, y_train)

y_pred_GPR = GPR.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Gaussian Process Regressor & TF-IDF: {accuracy_score(y_test, y_pred_GPR.round()) * 100 :.2f}%")
```

#### 11. Using `Ridge Classifier`

```python
from sklearn.linear_model import RidgeClassifier
RR = RidgeClassifier()
RR.fit(x_train, y_train)

y_pred_RR = RR.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Ridge Regression & TF-IDF: {accuracy_score(y_test, y_pred_RR) * 100:.2f}%")
```

#### 12. Using `Elastic Net Classifier`

```python
from sklearn.linear_model import ElasticNet
EN = ElasticNet()
EN.fit(x_train, y_train)

y_pred_EN = EN.predict(x_test).round()

print(f"Elastic Net Regression & TF-IDF: {accuracy_score(y_test, y_pred_EN) * 100 :.2f}%")
```

#### 13. Using `Multilayer Perceptron`

```python
from sklearn.neural_network import MLPClassifier
MLP = MLPClassifier(max_iter=10000)
MLP.fit(x_train, y_train)

y_pred_MLP = MLP.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"Multilayer Perceptron & TF-IDF: {accuracy_score(y_test, y_pred_MLP) * 100 :.2f}%")
```

#### 14. Using `Lasso Regressor`

```python
from sklearn.linear_model import Lasso
LASSO = Lasso()
LASSO.fit(x_train, y_train)

y_pred_LASSO = LASSO.predict(x_test)
y_pred_LASSO = [round(val) for val in y_pred_LASSO]

from sklearn.metrics import accuracy_score
print(f"Lasso Regression & TF-IDF: {accuracy_score(y_test, y_pred_LASSO) * 100 :.2f}%")
```

#### 15. Using `XG Boost`

```python
import xgboost as xgb
XGB = xgb.XGBClassifier()
XGB.fit(x_train, y_train)

y_pred_XGB = XGB.predict(x_test)

from sklearn.metrics import accuracy_score
print(f"XG Boost & TF-IDF: {accuracy_score(y_test, y_pred_XGB) * 100 :.2f}%")
```
